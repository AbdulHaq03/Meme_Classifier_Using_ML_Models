```python
df = pd.read_csv('labels.csv')
del df['text_ocr']
del df['Unnamed: 0']
df
```

```python
df['text_corrected'] = df['text_corrected'].str.replace('\d+', '')
df['text_corrected'] = df['text_corrected'].str.lower()
df['text_corrected'].astype(str).apply(lambda x: x.encode('latin-1', 'ignore').decode('latin-1'))
df['text_corrected'] = df['text_corrected'].str.replace('[^\w\s]','')
df['text_corrected'] = df['text_corrected'].str.replace('\s+',' ')
df['text_corrected'] = df['text_corrected'].astype(str)
df['text_corrected'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df
```

```python
import nlpaug.augmenter.word as naw
cleaned_text = []
for i in range(len(df['text_corrected'])):
    text = df.text_corrected[i]
    syn_aug = naw.SynonymAug(aug_src='wordnet',aug_max=4)
    syn_aug_text = syn_aug.augment(text,n=1)
    cleaned_text.append(''.join(syn_aug_text))
    
df['text_corrected'] = cleaned_text
df
```

```python
d={'very_positive':1,'positive':1, 'very_negative':-1, 'negative':-1, 'neutral':0}
df["overall_sentiment"]=df["overall_sentiment"].map(d)
X = df.text_corrected
y = df.overall_sentiment

le = LabelEncoder()
y = le.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 51)# 70% training and 30% test
```

```python
vectorizer = TfidfVectorizer(ngram_range = (1, 2))
X_trainn = vectorizer.fit_transform(X_train)
X_testt = vectorizer.transform(X_test)
```

```python
#DECISION TREE MODEL
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_trainn,y_train)
y_pred=clf.predict(X_testt)
print('Accuracy:', accuracy_score(y_test, y_pred)*100)
print("F1 Score: ",f1_score(y_test, y_pred,average='macro')*100) 
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
# Saving model
import pickle
Pkl_Filename = "Pickle_decision_tree_Model.pkl"  

with open(Pkl_Filename, 'wb') as file:  
    pickle.dump(clf, file)
```

```python
# Load the Model back from file
with open(Pkl_Filename, 'rb') as file:  
    Pickled_tree_Model = pickle.load(file)

Pickled_tree_Model
```

```python
# Calculate the Score 
score = Pickled_tree_Model.score(X_testt, y_test)  
# Print the Score
print("Test score: {0:.2f} %".format(100 * score))  

# Predict the Labels using the reloaded Model
Ypredict = Pickled_tree_Model.predict(X_testt)  

Ypredict
```

```python
clf = LogisticRegression(solver = 'newton-cg', class_weight = 'balanced')
clf.fit(X_trainn,y_train)
y_pred=clf.predict(X_testt)
print('Accuracy:', accuracy_score(y_test, y_pred)*100)
print("F1 Score: ",f1_score(y_test, y_pred,average='macro')*100) 
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
```
